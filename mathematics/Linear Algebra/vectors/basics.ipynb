```python
import matplotlib.pyplot as plt
import numpy as np

# starting points
X = [0, 0]
Y = [0, 0]

# vector components
U = [1, 2]
V = [2, 1]

plt.quiver(X, Y, U, V, angles='xy', scale=1, scale_units='xy',
           color=['r','b'], width=0.01, headwidth=5, pivot='tail',alpha=0.5)

plt.xlim(-1, 3)
plt.ylim(-1, 3)
plt.grid()
plt.show()
```

```python
# simple vector 

plt.grid()
plt.quiver(0, 0, 1, 2, angles='xy', scale_units='xy', scale=1,color='r')

plt.xlim(-1, 3)
plt.ylim(-1, 3)
plt.axhline(0, color='k', linewidth=0.5)
plt.axvline(0, color='k', linewidth=0.5)
plt.title("Vector (1,2)")
```

```python
v1 = np.array([1, 2])   
v2 = np.array([2, 1])  
v_sum = v1 + v2       

plt.figure(figsize=(12, 12))

plt.subplot(3,2,1)
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.title("Vector: (1,2)", fontsize=14, pad=15)
plt.xlim(0,4)
plt.ylim(0,4)
plt.grid()

plt.subplot(3,2,2)
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b')
plt.title("Vector: (2,1)", fontsize=14, pad=15)
plt.xlim(0,4)
plt.ylim(0,4)
plt.grid()

plt.subplot(3,2,3)
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label="(1,2)")
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label="(2,1)")
plt.title("Vectors: (1,2) & (2,1)", fontsize=14, pad=5)
plt.xlim(0,4)
plt.ylim(0,4)
plt.grid()
plt.legend()

plt.subplot(3,2,4)
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label="(1,2)")
plt.quiver(1, 2, 2, 1, angles='xy', scale_units='xy', scale=1, color='b', label="(2,1)")
plt.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='g', label="(3,3)")
plt.title("Vector Addition: (1,2) + (2,1) = (3,3)", fontsize=14, pad=5)
plt.xlim(0,4)
plt.ylim(0,4)
plt.grid()
plt.legend()

plt.tight_layout()
```

```python
v = np.array([2, 2])

plt.figure(figsize=(10, 8))

plt.subplot(2,2,1)
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.title("Vector: (2,2)", fontsize=14, pad=15)
plt.xlim(0,5)
plt.ylim(0,5)
plt.grid()

plt.subplot(2,2,2)
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label="(2,2)")
plt.quiver(0, 0, 2*v[0], 2*v[1], angles='xy', scale_units='xy', scale=1, color='g', label="2*(2,2)",alpha=0.4)
plt.title("Scalar Multiplication: 2*(2,2)", fontsize=14, pad=15)
plt.xlim(0,5)
plt.ylim(0,5)
plt.grid()
plt.legend()

plt.subplot(2,2,3)
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label="(2,2)")
plt.quiver(0, 0, 0.5*v[0], 0.5*v[1], angles='xy', scale_units='xy', scale=1, color='b', label="(1/2)*(2,2)",alpha=0.5)
plt.title("Scalar Multiplication: (1/2)*(2,2)", fontsize=14, pad=15)
plt.xlim(0,5)
plt.ylim(0,5)
plt.grid()
plt.legend()

plt.subplot(2,2,4)
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label="(2,2)")
plt.quiver(0, 0, 2*v[0], 2*v[1], angles='xy', scale_units='xy', scale=1, color='g', label="2*(2,2)",alpha=0.5)
plt.quiver(0, 0, 0.5*v[0], 0.5*v[1], angles='xy', scale_units='xy', scale=1, color='b', label="(1/2)*(2,2)",alpha=0.5)
plt.title("All in One", fontsize=14, pad=15)
plt.xlim(0,6)
plt.ylim(0,6)
plt.grid()
plt.legend()

plt.tight_layout()
```
