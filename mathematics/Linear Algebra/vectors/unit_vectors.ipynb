```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.quiver(0,0,0,1, angles='xy', scale_units='xy', scale=1, color='r')
plt.title("unit vector j hat", fontsize=14, pad=15)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.axhline(0, color='k', linewidth=0.5)
plt.axvline(0, color='k', linewidth=0.5)
plt.grid()

plt.tight_layout()
```

```python
plt.quiver(0,0,1,0, angles='xy', scale_units='xy', scale=1, color='r')
plt.title("unit vector i hat", fontsize=14, pad=15)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.axhline(0, color='k', linewidth=0.5)
plt.axvline(0, color='k', linewidth=0.5)
plt.grid()

plt.tight_layout()
```

```python
# unit vectors in 2D. i hat -> for x co-ordinate and j hat for y co-ordinate

plt.quiver(0,0,0,1, angles='xy', scale_units='xy', scale=1, color='r',label="j hat")
plt.quiver(0,0,1,0, angles='xy', scale_units='xy', scale=1, color='b',label="i hat")
plt.title("unit vectors in 2D", fontsize=14, pad=15)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.axhline(0, color='k', linewidth=0.5)
plt.axvline(0, color='k', linewidth=0.5)
plt.grid()
plt.legend()

plt.tight_layout()
```

```python
# unit vectors with scalars

plt.quiver(0,0,0,-2*1, angles='xy', scale_units='xy', scale=1, color='r',label="j hat")
plt.quiver(0,0,3*1,0, angles='xy', scale_units='xy', scale=1, color='b',label="i hat")
plt.title("unit vectors with scalars", fontsize=14, pad=15)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.axhline(0, color='k', linewidth=0.5)
plt.axvline(0, color='k', linewidth=0.5)
plt.grid()
plt.legend()

plt.tight_layout()
```

```python
# unit vectors with scalars

plt.quiver(0,0,0,-2*1, angles='xy', scale_units='xy', scale=1, color='r',label="j hat")
plt.quiver(0,0,3*1,0, angles='xy', scale_units='xy', scale=1, color='b',label="i hat")
plt.quiver(3,0,0,-2, angles='xy', scale_units='xy', scale=1, color='g',label="sum")
plt.quiver(0,0,3,-2, angles='xy', scale_units='xy', scale=1, color='g')
plt.title("unit vectors with scalars", fontsize=14, pad=15)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.axhline(0, color='k', linewidth=0.5)
plt.axvline(0, color='k', linewidth=0.5)
plt.grid()
plt.legend()

plt.tight_layout()
```
