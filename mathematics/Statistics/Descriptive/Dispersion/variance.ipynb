# Variance: σ2=N∑(xi​−μ)2​

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_point_1 = np.array([5, 4, 1, 6, 8, 9, 12, 15, 16, 18, 21, 22, 23, 24, 28, 29, 30]
)
data_point_2 = np.array([10, 200, 50, 300, 20, 150, 500, 5, 250, 400,
 30, 180, 60, 350, 12, 220, 90, 450, 40, 275,
 15, 300, 120, 380, 25, 170, 80, 420, 55, 260,
 35, 310, 100, 390, 45, 200, 70, 480, 65, 330]
)

mean1 = np.mean(data_point_1)
mean2 = np.mean(data_point_2)

print(mean1)
print(mean2)
```

```python
plt.scatter(data_point_1, data_point_1, marker = 'o', linestyle = '--', alpha = 0.6)
plt.plot(mean1, mean1, marker = 'o', linestyle = '--', alpha = 0.6, color = 'red', markersize = 10)
```

```python
plt.scatter(data_point_2, data_point_2, marker = 'o', linestyle = '--', alpha = 0.6)
plt.plot(mean2, mean2, marker = 'o', linestyle = '--', alpha = 0.6, color = 'red', markersize = 10)
```

```python
variance1 = np.sum(np.array([(i - mean1) ** 2 for i in data_point_1])) / len(data_point_1)
variance1 = np.sum(np.array([(i - mean1) ** 2 for i in data_point_1])) / len(data_point_1)
print('Variance1:', variance1)
# also show numpy's var for verification
print('np.var(data_point_1):', np.var(data_point_1))

# Visualization: show deviations from the mean as vertical lines
fig, ax = plt.subplots(figsize=(8, 4))
x = np.arange(len(data_point_1))
y = data_point_1
ax.scatter(x, y, alpha=0.7)
ax.hlines(mean1, -0.5, len(x)-0.5, colors='red', linestyles='--', label=f'Mean = {mean1:.2f}')
for xi, yi in zip(x, y):
    ax.vlines(xi, mean1, yi, colors='gray', alpha=0.6)
ax.set_xlabel('Index')
ax.set_ylabel('Value')
ax.set_title('Data Point 1 — deviations from mean (vertical lines show deviation)')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
variance2 = np.sum(np.array([(i - mean2) ** 2 for i in data_point_2])) / len(data_point_2)
print('Variance2:', variance2)
print('np.var(data_point_2):', np.var(data_point_2))

# Visualization for data_point_2 — many values, bigger figure
fig, ax = plt.subplots(figsize=(10, 5))
x2 = np.arange(len(data_point_2))
y2 = data_point_2
ax.scatter(x2, y2, alpha=0.6, s=20)
ax.hlines(mean2, -0.5, len(x2)-0.5, colors='red', linestyles='--', label=f'Mean = {mean2:.2f}')
for xi, yi in zip(x2, y2):
    ax.vlines(xi, mean2, yi, colors='gray', alpha=0.4)
ax.set_xlabel('Index')
ax.set_ylabel('Value')
ax.set_title('Data Point 2 — deviations from mean (vertical lines show deviation)')
ax.legend()
plt.tight_layout()
plt.show()
```
