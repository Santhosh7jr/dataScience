# Percentiles and Quartiles

```python
import numpy as np
```

```python
array = np.array([10, 20, 20, 30, 40, 50, 40, 30, 60, 70, 20, 10, 80, 90])
array = np.sort(array)
x = 80

number_of_values_below_x = 0

for i in array:
    if i >= x:
        break
    number_of_values_below_x += 1

percentile_of_x = number_of_values_below_x / len(array)

print(f"{format(percentile_of_x * 100, '.2f')}% of the distribution is less than {x}")
```

```python
first_quartile_upper_limit = (25 / 100) * (len(array) + 1)
second_quartile_upper_limit = (50 / 100) * (len(array) + 1)
third_quartile_upper_limit = (75 / 100) * (len(array) + 1)
fourth_quartile_upper_limit = (100 / 100) * (len(array) + 1)

print(f"{25}% of the distribution is less than {np.round(first_quartile_upper_limit)} that is {array[0 : int(first_quartile_upper_limit)]}")
print(f"{50}% of the distribution is less than {np.round(second_quartile_upper_limit)} that is {array[int(first_quartile_upper_limit) : int(second_quartile_upper_limit)]}")
print(f"{75}% of the distribution is less than {np.round(third_quartile_upper_limit)} that is {array[int(second_quartile_upper_limit) : int(third_quartile_upper_limit)]}")
print(f"{100}% of the distribution is less than {np.round(fourth_quartile_upper_limit)} that is {array[int(third_quartile_upper_limit) : int(fourth_quartile_upper_limit)]}")
```

# Inter Quartile Range

The Interquartile Range (IQR) is the range (difference) between:

Q3 → the 75th percentile

Q1 → the 25th percentile

### IQR = Q3 – Q1

It measures the spread of the middle 50% of your data.

|---- 25% (Q1) ----|-------- 50% (IQR) --------|---- 25% (Q3) ----|

```python
array = np.array([-10, -25, -50, -80, 1, 2, 2, 2, 3, 3, 4, 5, 5, 5, 6, 6, 6, 6, 7, 8, 8, 9, 27, 50, 75, 100])
array = np.sort(array)

q1 = array[int(np.round((25 / 100) * len(array)))]
q3 = array[int(np.round((75 / 100) * len(array)))]

iqr = q3 - q1

lower_fence = q1 - (1.5 * iqr)
upper_fence = q3 + (1.5 * iqr)

cleaned_data = []
outliers = []

for i in array:
    if i < lower_fence or i > upper_fence:
        outliers.append(int(i))
    else:
        cleaned_data.append(int(i))

print(f"Actual data: {array}")
print(f"cleaned Data: {cleaned_data}")
print(f"outliers: {outliers}")
```

# 5 Number Summary
min ---|--- Q1 ---[======= median =======]--- Q3 ---|--- max


```python
array = np.array([-10, -25, -50, -80, 1, 2, 2, 2, 3, 3, 4, 5, 5, 5, 6, 6, 6, 6, 7, 8, 8, 9, 27, 50, 75, 100])
array = np.sort(array)

minimum = array[0]
q1 = array[int(np.round((25 / 100) * len(array)))]
median = np.median(array)
q3 = array[int(np.round((75 / 100) * len(array)))]
maximum = array[len(array) - 1]

print(f"{minimum} ---|--- {q1} ---[======= {median} =======]--- {q3} ---|--- {maximum}")
```
