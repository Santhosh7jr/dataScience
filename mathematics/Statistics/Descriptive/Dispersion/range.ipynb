```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
year_wise_marks = np.array([80, 76, 85, 100, 100, 100], dtype=int)
min = np.min(year_wise_marks)
max = np.max(year_wise_marks)

rang = max - min

rang
```

```python
plt.plot(rang, rang, color = 'red', marker='o', markersize = 10, label='marker')
plt.plot(year_wise_marks, year_wise_marks, marker = 'o', linestyle='--', label='dataPoints', alpha=0.5)
```

```python
year_wise_marks[0] = 0
min = np.min(year_wise_marks)
max = np.max(year_wise_marks)

rang = max - min

rang
```

```python
plt.plot(rang, rang, color = 'red', marker='o', markersize = 10, label='marker')
plt.plot(year_wise_marks, year_wise_marks, marker = 'o', linestyle='--', label='dataPoints', alpha=0.5)
```
