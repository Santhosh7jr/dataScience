# Covariance => Cov(X,Y)=n1​∑(Xi​−Xˉ)(Yi​−Yˉ)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
# Positive Covariance

hours_studied = np.array([2, 3, 4, 5, 6])
marks = np.array([50, 60, 70, 80, 90])

study_mean = np.mean(hours_studied)
marks_mean = np.mean(marks)
```

```python
covariance = np.sum([(((hours_studied[i] - study_mean) * (marks[i] - marks_mean))) for i in range(len(hours_studied))]) / (len(hours_studied) - 1)
covariance
```

```python
covariance = np.cov(hours_studied, marks)
covariance
```

```python
# Negative covariance

x = [4, 5, 6, 7, 8, 9]
y = [24, 12, 6, 3, 2, 1]

np.cov(x, y)[1][0]
```

```python
# Zero covariance

x = [1, 6, 8, 9, 10, 11]
y = [4, 2, 5, 4, 8, 1]

np.cov(x, y)[1][0]
```

# Correlation

How much X and Y move together
divided by
how much they move individually

# Pearson Correlation (r)
✔ Measures linear relationship

✔ Works with actual values

✔ Sensitive to outliers

✔ Assumes the data is continuous and normally distributed## Pearson Correlation Coefficient

```python
scipy.stats.pearsonr(hours_studied, marks)
```

```python
scipy.stats.pearsonr(x, y)
```

# Spearman Correlation (ρ)

✔ Measures monotonic relationship
(monotonic = always increasing or always decreasing, but not necessarily linearly)

✔ Works on ranks instead of values

✔ Not sensitive to outliers

✔ Works for non-normal data

```python
scipy.stats.spearmanr(hours_studied, marks)
```
