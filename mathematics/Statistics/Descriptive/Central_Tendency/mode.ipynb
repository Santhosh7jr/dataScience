```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([1, 2, 3, 3, 2, 1, 6, 7, 8, 9000, 1, 2, 3, 4, 5, 1, 2])
data = np.sort(data)
```

```python
freqDict = {}
maximum = 0

for i in data:
    freqDict[i] = freqDict.get(i, 0) + 1
    if freqDict[i] > maximum:
        maximum = freqDict[i]
        
mode = []

for key, val in freqDict.items():
    if val == maximum :
        mode.append(key)

mode
```

```python
plt.plot(mode, mode, color = 'red', marker='o', markersize = 10, label='marker')
plt.plot([i for i in range(1, len(data) + 1)], data, marker = 'o', linestyle='--', label='dataPoints', alpha = 0.5, color = 'blue')
```
