```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(np.arange(0, 3), [np.log(i) for i in range(0, 3)], marker="o")
plt.plot(np.arange(2, 5), [(i ** 2) * np.log(i) for i in range(2, 5)], marker="o")
plt.grid()
```

```python

v1 = np.array([1, 2])
v2 = np.array([3, 1])


plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1)
for i in range(-30, 30):
    c = i * 0.1
    plt.quiver(
        v1[0], v1[1],
        v2[0] * c,
        v2[1] * c,
        angles='xy', scale_units='xy', scale=1, color='r'
    )

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid()

```

```python
plt.plot(np.arange(0, 20), [-((i * 0.1) / 3 * (np.sin(i * 0.1 - 2))) for i in range(0, 20)], marker="o")
plt.plot(np.arange(21, 30), [-((i * 0.1) / 3 * (np.sin(i * 0.1 - 2) )) for i in range(21, 30)], marker="o")
plt.grid()
```
