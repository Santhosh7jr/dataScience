```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [i for i in range(-5, 5)]
y1 = [((6 * (i ** 2)) + (5 * i) - 1) for i in x] # 6x^2 + 5x - 1
y2 = [((-3 * (i ** 3)) - (7 * i) + 8) for i in x] # -3x^3 - 7x + 8

plt.plot(x, y1, marker="o")
plt.plot(x, y2, marker="o")
plt.grid()
plt.axhline(0)
plt.axvline(0)
plt.grid(True)
```

```python
y3 = [(i / np.log(i)) for i in range(2, 10)]

plt.plot([i for i in range(2, 10)], y3, marker="o")
plt.axhline(0)
plt.axvline(0)
plt.grid(True)
plt.show()
```

```python
y1 = [((i + 2) / (i - 1)) for i in range(0, 5) if i != 1]
y1.insert(1, -20)

y2 = [np.sqrt(i) for i in range(4, 10)]

plt.plot(np.arange(0, 5), y1, marker="o")
plt.plot(np.arange(4, 10), y2, marker="o")
plt.grid()
```

```python
y1 = [(np.sin((i + 1) * (np.pi / 180))) for i in range(-5, 0)]
y2 = [(2 ** i) for i in range(-1, 4)]

print(y1)
print(y2)

plt.plot(np.arange(-5, 0), y1, marker="o")
plt.plot(np.arange(-1, 4), y2, marker="o")
plt.grid()
```

```python
y1 = [((i ** 2) - 1) for i in range(-5, 4)]
y2 = [((2 * i) + 1) for i in range(3, 10)]

print(y1)
print(y2)

plt.plot(np.arange(-5, 4), y1, marker="o")
plt.plot(np.arange(3, 10), y2, marker="o")
plt.grid()
```

```python
y1 = [(1 / i) for i in range(-5, 0)]
y2 = [((2 * i) + 1) for i in range(-1, 5)]

print(y1)
print(y2)

plt.plot(np.arange(-5, 0), y1, marker="o")
plt.plot(np.arange(-1, 5), y2, marker="o")
plt.grid()
```

```python
y1 = [((2 * i) + 1) for i in range(-5, -1)]
y2 = [(1 /(1 - i)) for i in range(-2, 1)]

print(y1)
print(y2)

plt.plot(np.arange(-5, -1), y1, marker="o")
plt.plot(np.arange(-2, 1), y2, marker="o")
plt.grid()
```

```python
y1 = [((i - 3) * (i + 4)) / (i + 5) for i in range(-10, -5)]
y2 = [((i - 3) * (i + 4)) / (i + 5) for i in range(-4, 1)]

print(y1)
print(y2)

plt.plot(np.arange(-10, -5), y1, marker="o")
plt.plot(np.arange(-4, 1), y2, marker="o")
plt.grid()
```
