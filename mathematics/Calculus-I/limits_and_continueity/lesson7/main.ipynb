```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y1 = [((i + 1) / ((np.sqrt(i + 5)) - 2)) for i in range(-10, 10) if i != -1]
y1.insert(9, -10)
y2 = [(np.sqrt(i + 5) + 2) for i in range(-10, 10)]

print(y1)
print(y2)

plt.plot(np.arange(-10, 10), y1, marker="o")
plt.plot(np.arange(-10, 10), y2, marker="o")
plt.grid()
```

```python
y = [((1 - (np.cos(i))) / 2 * np.sin(i)) for i in range(-20, 21) if i != 0]
y.insert(20, (1 / (2 * (1 + np.cos(0)))))

plt.plot(np.arange(-20, 21), y, marker="o")
plt.grid()
```
