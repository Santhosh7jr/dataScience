## LCM and GCD

```python
def find_GCD(a, b):

    if b == 0:
        return a
    return find_GCD(b, a % b)


find_GCD(1, 5)
```

```python
# GCD(a, b) = GCD(b, a) 
print(find_GCD(5, 7))
print(find_GCD(7, 5))
```

```python
def demo_GCD():

    a = int(input("Enter the value of 'a': "))
    b = int(input("Enter the value of 'b': "))

    arr = set()
    result = find_GCD(a, b)

    print(result)

    for i in range(0, 5):
        for j in range(0, 5):
            arr.add(((a * i) + (b * j)) % result)

    return arr

demo_GCD()
```

```python
def find_LCM(a, b):
    
    return int((a / find_GCD(a, b)) * b)
```

```python
def demo_LCM(a, b):

    arr = set()
    result = find_LCM(a, b)

    print(result)

    for i in range(0, 5):
        for j in range(0, 5):
            arr.add(((a * i) + (b * j)))

    print(arr)

    return result
```

```python
def find_factors(a):

    result = [1]

    factor = 2

    while(a != 1):

        if (a % factor) == 0:
            a = int(a / factor)
            result.append(factor)
        else:
            factor += 1

    return result

a = int(input("Enter the value of 'a': "))
b = int(input("Enter the value of 'b': "))

    
result = demo_LCM(a, b)
find_factors(a), find_factors(b), find_factors(result)
```
