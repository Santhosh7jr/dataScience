```python
import sqlite3
```

```python
connection = sqlite3.connect("example.db")
connection
```

```python
cursor = connection.cursor();
```

```python
cursor.execute("""
    create table if not exists student(
        id int primary key,
        name text not null,
        age int      
    );
""")

connection.commit()
```
