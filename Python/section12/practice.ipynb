## Iterators

```python
iterator = iter([1, 2, 3, 4, 5])
print(iterator)
```

```python
for item in iterator:
    print(item)
    break
```

```python
next(iterator)
```

## Generators

```python
def square(n):
    for i in range(1, n + 1):
        return i ** 2
    
print(square(5))
```

```python
def square(n):
    for i in range(1, n + 1):
        yield i ** 2

print(square(5))
```

```python
generator = square(5)
```

```python
next(generator)
```

```python
def readFile(filename):
    with open(filename, 'r') as file:
        for line in file.readlines():
            yield line

object = readFile("sample.txt")
```

```python
for line in object:
    print(line)
```
