# understanding equation y=mx+c

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
import time
```

```python
def line_equation(m,x,c):
    return (m*x)+c # <-- straight line equation

x=[i for i in range(25)]
m=1
c=2 # <-- always constant

y=[line_equation(m,i,c) for i in x]

plt.grid()
plt.plot(x,y)
```

```python
def line_equation(m,x,c):
    return (m*x)+c

x=[i for i in range(25)]
m=[(i*0.001) for i in x] # <-- if slope varies then its not a straight line
c=2

y=[line_equation(m[i],x[i],c) for i in range(len(x))]

plt.grid()
plt.plot(x,y,marker='o')
```

```python
def line_equation(m,x,c):
    return (m*x)+c

x=[i for i in range(25)]
m=[(i%2) for i in x] # <-- if slope varies then its not a straight line
c=0

y=[line_equation(m[i],x[i],c) for i in range(len(x))]

plt.grid()
plt.plot(x,y,marker='o')
plt.xlim(0,25)
plt.ylim(0,100)
```

```python
def line_equation(m,x,c):
    return (m*x)+c

x=[i for i in range(25)]
m=2
c=0

y=[line_equation(m,x[i],c) for i in range(len(x))]

plt.grid()
plt.plot(x,y,marker='o')
plt.xlim(0,25)
plt.ylim(0,100)
```

```python
c = 0
x = [i for i in range(0, 11)]
m = 0
xlist = []
ylist = []

for i in range(0, 11):
    y = [(m * i) + c for i in x]
    c += 1
    xlist.append(x)
    ylist.append(y)
    for i in range(len(xlist)):
        plt.plot(xlist[i], ylist[i], marker='o')
    plt.show()
    time.sleep(1)
```

```python
import matplotlib.pyplot as plt

x = [i for i in range(-10, 11)]
m1 = 0
m2 = 0
count = 1
xlist = []
ylist = []

plt.ion()

for i in range(0, 10):

    if count == 360:
        break

    y1 = [(m1 * i) for i in x]
    y2 = [(m2 * i) for i in x]

    m1 += 30
    m2 -= 30

    ylist.append(y1)
    ylist.append(y2)

    plt.clf()
    plt.grid(True)

    for j in range(len(ylist)):
        plt.plot(x, ylist[j])

    plt.grid(True)

    plt.xlim(-15, 15)

    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1) 

    plt.show()
    plt.pause(1)
```
