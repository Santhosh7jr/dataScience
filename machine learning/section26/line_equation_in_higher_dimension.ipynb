```python
import matplotlib.pyplot as plt

w = [2, 0]
x = [0.3, 2]

plt.figure(figsize=(6,6))

# Axis limits
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Plot vector w
plt.quiver(
    0, 0, 
    w[0], w[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    color='r',
    width=0.01,
    headwidth=5,
    pivot='tail'
)

# Plot vector x
plt.quiver(
    0, 0,
    x[0], x[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    color='b',
    width=0.01,
    headwidth=5,
    pivot='tail'
)

# Axes & grid
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid()
plt.gca().set_aspect('equal')

plt.show()

```

```python
res = sum([w[i] * x[i] for i in range(len(w))])
magn = np.sqrt(sum([i * i for i in x]))

res/magn
```
