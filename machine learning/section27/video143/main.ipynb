# Cost Function

```python
import matplotlib.pyplot as plt
import time
```

```python
x = [i for i in range(1, 6)]
x
```

```python
for i in range(-3, 3):

    output = []
    result = 0

    for inp in x:
        output.append(i * inp)
    
    plt.plot(x, output, marker = 'o', label = f"slope = {i}")

    for j in range(len(x)):
        result += (output[j] - x[j]) ** 2

    print(f"slope = {i}, result = {result / (2 * len(x))}")

plt.legend()
plt.grid()
plt.xlim(0, 10)
plt.ylim(-15, 15)
```
