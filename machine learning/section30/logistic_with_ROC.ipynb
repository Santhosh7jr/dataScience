```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, RandomizedSearchCV
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from collections import Counter
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
```

```python
x, y = make_classification(n_samples=1000, n_classes=2, random_state=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
dummy_model = [0 for _ in range(len(y_test))]
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model_proba = model.predict_proba(x_test)
```

```python
model_proba = model_proba[:, 1]
```

```python
dummy_model_auc = roc_auc_score(y_test, dummy_model)
model_auc = roc_auc_score(y_test, model_proba)
dummy_model_auc, model_auc
```

```python
dummy_fpr, dummy_tpr, _ = roc_curve(y_test, dummy_model)
model_fpr, model_tpr, _ = roc_curve(y_test, model_proba)
```

```python
plt.plot(dummy_fpr, dummy_tpr, linestyle='--')
plt.plot(model_fpr, model_tpr, marker='.')
```
