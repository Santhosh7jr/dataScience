```python
import numpy as np
import pandas as pd
```

```python
matrix = [[80,-5,60],[78,-3,62],[82,-6,65],[81,-4,68]]

matrix
```

```python
df = pd.DataFrame(matrix,columns=['N1','N2','N3'], index=['A1','A2','A3','A4'])
df
```

```python
maxi_max = []
maxi_min = []
equally_likely = []
hurwicz = []
minimax_regret = []
max_num = []

alpha = 0.3

for i in matrix :
    maxi_max.append(max(i))
    maxi_min.append(min(i))
    equally_likely.append(sum(i)/len(i))
    hurwicz.append((alpha*max(i))+((1 - alpha)*min(i)))

for i in df.columns:
    max_num.append(max(df[i]))

regret_matrix = [[max_num[j] - matrix[i][j] for j in range(len(matrix[i]))] for i in range(len(matrix))]

for i in regret_matrix:
    minimax_regret.append(max(i))

df['maxi_max'] = maxi_max
df['maxi_min'] = maxi_min
df['equally_likely'] = equally_likely
df['hurwicz'] = hurwicz
df['minimax_regret'] = minimax_regret

df.loc['Result'] = [None] * df.shape[1]
df
```

```python
print(f"maxi_max --> A{df.index[df['maxi_max'] == max(df['maxi_max'])]}")
print(f"maxi_min --> A{df.index[df['maxi_min'] == max(df['maxi_min'])]}")
print(f"equally_likely --> A{df.index[df['equally_likely'] == max(df['equally_likely'])]}")
print(f"hurwicz --> A{df.index[df['hurwicz'] == max(df['hurwicz'])]}")
print(f"minimax_regret --> A{df.index[df['minimax_regret'] == min(df['minimax_regret'])]}")
```
